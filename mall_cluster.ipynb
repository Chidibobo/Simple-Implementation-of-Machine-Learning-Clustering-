Importing libraries for Data Analysis and Visualization

```python
#import libraries required
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
```

KMeans Clustering Models

```python
#read Mall_Customer file 
cluster_dataset = pd.read_csv('Mall_Customers.csv')
#create an array with features of Annual Income(k$) and Spending Score(1-100)
x_array = cluster_dataset.iloc[:,[3,4]].values
x_array
```

```python
# using The Elbow method using the 'Within cluster sum of Squares'

#import Kmeans algorithm from the sklearn library
from sklearn.cluster import KMeans
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters= i, init= 'k-means++' , random_state= 50)
    kmeans.fit(x_array)
    wcss.append(kmeans.inertia_)

#plotting the Graph to visualize the optimal Number of Clusters
plt.plot(range(1,11) , wcss)
plt.title('ELBOW METHOD TO DETERMINE N OF CLUSTERS')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Giving each instance of the dataset a cluster number
kmeans = KMeans(n_clusters= 5 , init ='k-means++' , random_state= 50)
y_kmeans = kmeans.fit_predict(x_array)
y_kmeans
```

```python
#Plotting the scatter plot to visualize the Kmeans Model

plt.scatter(x_array[y_kmeans == 0 , 0], x_array[y_kmeans == 0,1], s = 100, c = 'blue', label = 'Regular')
plt.scatter(x_array[y_kmeans == 1 , 0], x_array[y_kmeans == 1,1], s = 100, c = 'red', label = 'Priority')
plt.scatter(x_array[y_kmeans == 2 , 0], x_array[y_kmeans == 2,1], s = 100, c = 'green', label = 'Caution')
plt.scatter(x_array[y_kmeans == 3 , 0], x_array[y_kmeans == 3,1], s = 100, c = 'yellow', label = 'Target')
plt.scatter(x_array[y_kmeans == 4 , 0], x_array[y_kmeans == 4,1], s = 100, c = 'black', label = 'No Action')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 150, c = 'orange', label = 'Centriods')
plt.title('CLUSTERS OF CUSTOMERS')
plt.xlabel('ANNUAL INCOME (k$)')
plt.ylabel('SPENDING SCORE (1-100)')
plt.legend()
plt.show()
```

```python
#Evaluating the Silhouette Score of the Kmeans Model
from sklearn.metrics import silhouette_score
sil_score = silhouette_score(x_array, kmeans.labels_, metric ='euclidean')
print('Silhouette Score: %.3f' % sil_score)
```

```python
#read Mall_Customer file 
agecluster_dataset = pd.read_csv('Mall_Customers.csv')

#create an array with features of Age and Spending Score(1-100)
agex_array = agecluster_dataset.iloc[:,[2,4]].values
agex_array
```

```python
# using The Elbow method using the 'Within cluster sum of Squares'

#import Kmeans algorithm from the sklearn library
from sklearn.cluster import KMeans
wcssage = []
for i in range (1,11):
    kmeans = KMeans(n_clusters= i, init= 'k-means++' , random_state= 50)
    kmeans.fit(agex_array)
    wcssage.append(kmeans.inertia_)

#plotting the Graph to visualize the optimal Number of Clusters
plt.plot(range(1,11) , wcssage)
plt.title('ELBOW METHOD TO DETERMINE N OF CLUSTERS')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Giving each instance of the dataset a cluster number
kmeans = KMeans(n_clusters= 4 , init ='k-means++' , random_state= 50)
agey_kmeans = kmeans.fit_predict(agex_array)
agey_kmeans
```

```python
#Plotting the scatter plot to visualize the Kmeans Model

plt.scatter(agex_array[agey_kmeans == 0 , 0], agex_array[agey_kmeans == 0,1], s = 100, c = 'blue', label ='Target Customer(Old)')
plt.scatter(agex_array[agey_kmeans == 1 , 0], agex_array[agey_kmeans == 1,1], s = 100, c = 'red', label = 'Priority Customer(Young)')
plt.scatter(agex_array[agey_kmeans == 2 , 0], agex_array[agey_kmeans == 2,1], s = 100, c = 'green', label = 'Caution Customer(Across all)')
plt.scatter(agex_array[agey_kmeans == 3 , 0], agex_array[agey_kmeans == 3,1], s = 100, c = 'yellow', label = 'Target Customer(Young)')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 150, c = 'orange', label = 'Centriods')
plt.title('CLUSTERS OF CUSTOMERS BY AGE')
plt.xlabel('AGE')
plt.ylabel('SPENDING SCORE (1-100)')
plt.legend()
plt.show()
```

```python
#Evaluating the Silhouette Score of the Kmeans Model
from sklearn.metrics import silhouette_score
sil_score = silhouette_score(agex_array, kmeans.labels_, metric ='euclidean')
print('Silhouette Score: %.3f' % sil_score)
```

Hierarchical Clustering Models

```python
#read Mall_Customer file 
h_dataset = pd.read_csv('Mall_Customers.csv')

#create an array with features of Annual Income(k$) and Spending Score(1-100)
h_x= h_dataset.iloc[:, [3,4]].values
h_x
```

```python
#import the Hierarchy Clustering Algorithm from Scipy
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(h_x, method ='ward'))

#Visualize the Dendogram to evaluate the optimal number of clusters 
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
#import the Agglomerative Clustering alogrithm from sklearn
from sklearn.cluster import AgglomerativeClustering

#Giving each instance of the dataset a cluster number
hc = AgglomerativeClustering(n_clusters= 5 , affinity= 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(h_x)
y_hc
```

```python
#Plotting the scatter plot to visualize the Heirachical Clustering Model
plt.scatter(h_x[y_hc == 0 , 0], h_x[y_hc == 0,1], s = 100, c = 'blue', label = 'Target')
plt.scatter(h_x[y_hc == 1 , 0], h_x[y_hc == 1,1], s = 100, c = 'red', label = 'Regular')
plt.scatter(h_x[y_hc == 2 , 0], h_x[y_hc == 2,1], s = 100, c = 'green', label = 'Priority')
plt.scatter(h_x[y_hc == 3 , 0], h_x[y_hc == 3,1], s = 100, c = 'yellow', label = 'Caution')
plt.scatter(h_x[y_hc == 4 , 0], h_x[y_hc == 4,1], s = 100, c = 'black', label = 'No Action')
plt.title('CLUSTERS OF CUSTOMERS')
plt.xlabel('ANNUAL INCOME (k$)')
plt.ylabel('SPENDING SCORE (1-100)')
plt.legend()
plt.show()
```
